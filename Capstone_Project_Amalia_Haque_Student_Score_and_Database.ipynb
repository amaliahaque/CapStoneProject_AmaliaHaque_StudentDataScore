```python


# Student Database Dictionary
listStudent =  [
    {
        'student name': 'Amalia',
        'student id'  : 'BootCamp001',
        'program 1'   : 'Data Science',
        'score'       : 0
    },
    {
        'student name': 'Barack Obama',
        'student id'  : 'BootCamp002',
        'program 1'   : 'Data Science',
        'score'       : 0
     
    }
]

# [Database Menu for Admin] Code for showing for database Admin

def menuDatabaseAdmin():
  print("1. View Student Databases")
  print("2. Add Student Database")
  print("3. Delete Student Database")

# [Database Menu for Admin] code for showing student database

def viewStudentDatabases ():
  print('Student Databases\n')
  print('Index\t|Student Name\t|Student ID \t|Program 1\t|score')
  for i in range (len(listStudent)):
    print('{}\t| {}\t| {}\t| {}\t| {}'.format(i,listStudent[i]['student name'],listStudent[i]['student id'],listStudent[i]['program 1'],listStudent[i]['score']))

# [Database Menu for Admin] code for adding new student database
def addStudent() :
    namaStudent = input('Entry Student Name : ')
    studentId = input('Entry Student ID : ')
    program1  = input('Entry Program 1 : ')
    listStudent.append({
        'student name': namaStudent,
        'student id': studentId,
        'program 1': program1
    })
    print("Your New Entry has Sucessfully Saved!")

    print("This is The Updated Databases")

    viewStudentDatabases()

# [Database Menu for Admin] code for deleting student database
def deleteStudent():
  viewStudentDatabases()
  indexStudent = int(input('Entry the Index you want to delete : '))
  del listStudent[indexStudent]

  print("Your New Entry has Successfully Saved! ")

  print("This is The Updated Databases")

  viewStudentDatabases()

# [Report menu for Admin] code for menu for admin report
def menuReportAdmin():
  print("1. Entry or Edit Report")
  print("2. View Report")

# [Report menu for Admin] code for entry exam score for student  
def entryScore ():
  viewStudentDatabases()
  while True :
    entryIndex = int(input('Entry the Index to Update The Score: '))
    score1 = int(input('Entry the Score : '))
    listStudent[entryIndex]['score'] = (score1)
    checker = input('Update score for other student? (Yes/No) = ')
    if checker == 'Yes':
      print(viewStudentDatabases())
    elif checker == 'No':
      print(viewStudentDatabases())
      break

#[Report Menu for Admin] code for reviewing student report
def viewReport():
  viewStudentDatabases()

# [Database Menu for Admin] code for selecting student database menu
def studentDatabasesMenu ():
    menuDatabaseAdmin ()
    while True :
      selectedMenuDatabaseAdmin = input (''' Please Select Student Database Menu : ''')
      if selectedMenuDatabaseAdmin == '1':
        viewStudentDatabases()
      elif selectedMenuDatabaseAdmin == '2':
        addStudent()
      elif selectedMenuDatabaseAdmin == '3' :
        deleteStudent()
        break

# [Report Menu for Admin] code for selecting report admin menu
def reportAdminMenu ():
  menuReportAdmin()
  while True :
    selectedMenuReportAdmin = input (''' Please Select Menu Report Admin : ''')
    
    if selectedMenuReportAdmin == '1':
      entryScore()
    elif selectedMenuReportAdmin == '2':
      viewReport()
      break

# [Report Menu for Admin] code for showing admin menu
def adminMenu () :
  print("1. Student Databases")
  print("2. Student Report")

# [Report Menu for Admin] code for selecting report menu
def adminPage ():
  adminMenu()
  while True :
    selectedAdminMenu= input('''Please Select Admin Menu : ''')
    if selectedAdminMenu == '1':
      studentDatabasesMenu()
    elif selectedAdminMenu == '2':
      reportAdminMenu()
      continue

# [Student Page] Exam Result Announcement
def studentPage():
  entryStudentId = (input("Please input your student id to access the announcement : "))
  print('Exam Result Announcement!')
  print('Index\t|Student Name\t|Student ID \t|Program 1\t|score')
  for i in range (len(listStudent)):
    print('{}\t| {}\t| {}\t| {}\t| {}'.format(i,listStudent[i]['student name'],listStudent[i]['student id'],listStudent[i]['program 1'],listStudent[i]['score']))

#This is the code for the Main Menu

while True :
  welcomePage = input(''' 

  Welcome to Purwadhika Dashboard!

  Main Menu :
  1. Entry Admin
  2. Entry Student
  3. Exit 
  
  Select Main Menu : ''')

  if(welcomePage == '1') :
    adminPage()
  elif(welcomePage == '2') :
    studentPage()
  elif (welcomePage =='3'):
    break

```
